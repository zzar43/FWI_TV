```julia
using PyPlot, JLD2
```

```julia

@load "data_tv_1e-16.jld2" xk
xk_tv1 = copy(xk)

@load "data_tv_8e-16.jld2" xk
xk_tv2 = copy(xk)

@load "data_gn.jld2" c c_true xk
xk_gn = copy(xk);
```

```julia
Nx = 61
Ny = 101;
```

```julia
# plot model
ma = maximum(c_true)
mi = minimum(c_true)
figure(figsize=(10,5))

subplot(221)
imshow(c_true, cmap="viridis_r"); title("True velocity model"); colorbar(shrink=0.3)

subplot(222)
imshow(reshape(xk_gn,Nx,Ny), cmap="viridis_r"); title("Result without regularization"); colorbar(shrink=0.3); clim(mi, ma)

subplot(223)
imshow(reshape(xk_tv1,Nx,Ny), cmap="viridis_r"); colorbar(shrink=0.3); clim(mi, ma)
title("TV: lambda = 1e-16")

subplot(224)
imshow(reshape(xk_tv2,Nx,Ny), cmap="viridis_r"); colorbar(shrink=0.3); clim(mi, ma)
title("TV: lambda = 8e-16")

tight_layout()
```
